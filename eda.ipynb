```
def set_dtypes(df, cols):
    for c in cols:
        if df[c].max() < 256 and df[c].min() >= 0:
            df[c] = df[c].astype(np.uint8)
        elif df[c].max() < 65536 and df[c].min() >= 0:
            df[c] = df[c].astype(np.uint16)
        elif df[c].max() < 42949672956 and df[c].min() >= 0:
            df[c] = df[c].astype(np.uint32)
        elif df[c].max() < 128 and df[c].min() >= -128:
            df[c] = df[c].astype(np.int8)
        elif df[c].max() < 32768 and df[c].min() >= -32768:
            df[c] = df[c].astype(np.int16)
        else:
            df[c] = df[c].astype(np.int64)
```

# Feature Engineering 

Creating a new column that says whether or not a flight was delayed upon departure
- 0 means it is not delayed
- 1 means it is delayed

```
conditions = [(df['DEP_DELAY'] > 0 )]
choices = [1]
df['DEP_DELAY_Y/N'] = np.select(conditions, choices, 0)
```

Creating a new column that says whether or not a flight was delayed upon arrival
- 0 means it is not delayed
- 1 means it is delayed

```
conditions = [(df['ARR_DELAY'] > 0 )]
choices = [1]
df['ARR_DELAY_Y/N'] = np.select(conditions, choices, 0)
df.head() 
```

Create new columns for:
- year
- month
- day of month
- day of week
- time of day of departure (morning, afternoon, evening, night)

```
df['YEAR'] = pd.DatetimeIndex(df['FL_DATE']).year
df['MONTH'] = pd.DatetimeIndex(df['FL_DATE']).month
df['DAY_OF_MONTH'] = pd.DatetimeIndex(df['FL_DATE']).day
df['FL_DATE'] =  pd.to_datetime(df['FL_DATE'])
df['DAY_OF_WEEK'] = df['FL_DATE'].dt.day_name()
df.head()
```

```
def time_of_day(row):
  if row['DEP_TIME'] >= 500 and row['DEP_TIME'] < 1200:
    return 'morning'
  elif row['DEP_TIME'] >= 1200 and row['DEP_TIME'] < 1700:
    return 'afternoon'
  elif row['DEP_TIME'] >= 1700 and row['DEP_TIME'] < 2030:
    return 'evening'
  elif row['DEP_TIME'] >= 2030 or row['DEP_TIME'] < 500:
    return 'night'
  else:
    return 'night'
```

```
df['TIME_OF_DAY'] = df.apply(time_of_day, axis=1)
```

```
delay_df = df[df['DEP_DELAY'] > 0]
not_delay_df = df[df['DEP_DELAY'] <= 0]
```

```
def make_heatmap(x, y):
  heatmapdf = pd.pivot_table(delay_df[[y, x, 'DEP_DELAY']], index=[y, x], aggfunc='count')
  totaldf = pd.pivot_table(df[[y, x, 'DEP_DELAY']], index=[y, x], aggfunc='count')
  heatmapdf = heatmapdf.unstack()
  totaldf = totaldf.unstack()
  heatmap_normdf = heatmapdf / totaldf *100

  plt.figure(figsize=(16,10))
  ax = sns.heatmap(heatmap_normdf, linewidths=.5, xticklabels=range(1,df[x].nunique()+1), cmap="bwr")
  plt.ylabel(y, fontsize=18)
  plt.xlabel(x,fontsize=18)
  plt.xticks(fontsize=12)
  plt.yticks(fontsize=12)
  plt.title('Delays by Date', fontsize=20);
```

```
make_heatmap(x='DAY_OF_MONTH', y='MONTH')
```

```
make_heatmap(x='MONTH', y='YEAR')
```

```
def make_heatmap_year(year, yeartitle):
  df_year = df[df['YEAR']==year]
  delay_df_year = df_year[df_year['DEP_DELAY'] > 0]
  not_delay_df_year = df_year[df_year['DEP_DELAY'] <= 0]
  heatmapdf = pd.pivot_table(delay_df_year[['MONTH', 'DAY_OF_MONTH', 'DEP_DELAY']], 
                            index=['MONTH', 'DAY_OF_MONTH'], aggfunc='count')
  totaldf = pd.pivot_table(df_year[['MONTH', 'DAY_OF_MONTH', 'DEP_DELAY']], index=['MONTH', 'DAY_OF_MONTH'], aggfunc='count')
  heatmapdf = heatmapdf.unstack()
  totaldf = totaldf.unstack()
  heatmap_normdf = heatmapdf / totaldf *100

  plt.figure(figsize=(16,10))
  ax = sns.heatmap(heatmap_normdf, linewidths=.5, xticklabels=range(1,32), cmap="bwr")
  plt.ylabel('Month', fontsize=18)
  plt.xlabel('Day',fontsize=18)
  plt.xticks(fontsize=12)
  plt.yticks(fontsize=12)
  plt.title('Delays by Date in '+ yeartitle, fontsize=20);
```
